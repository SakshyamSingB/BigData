1. Display Fibonacci series up to 10 terms.






```python
number_of_terms = 10
first_number = 0
second_number = 1

print("Fibonacci series:")

#calculating  fibonnaci series
for count in range(number_of_terms):
    print(first_number, end=" ")

    next_number = first_number + second_number
    first_number = second_number
    second_number = next_number
```

2. Find the factorial of a given number. Ask the user to provide the number.


```python
number = int(input("Enter a number: "))

factorial = 1

# calculate factorial
for i in range(1, number + 1):
    factorial = factorial * i

print("Factorial of", number, "is", factorial)
```

3. Write a program to print a right angle triangle using the for loop

```python
rows = 5

# outer loop controls rows
for i in range(1, rows + 1):
    # inner loop controls columns
    for j in range(i):
        print("*", end=" ")
    print()
```

1. Make a simple calculator that can add, subtract, multiply and divide two numbers using function.

```python
# for adding two numbers
def add(num1, num2):
    return num1 + num2

# for subtracting two numbers
def subtract(num1, num2):
    return num1 - num2

#  for multiplying two numbers
def multiply(num1, num2):
    return num1 * num2

# for divide=ing two numbers
def divide(num1, num2):
    if num2 == 0:
        return "Error! Division by zero"
    return num1 / num2

number1 = float(input("Enter first number: "))
number2 = float(input("Enter second number: "))

print("Choose operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = input("Enter choice (1/2/3/4): ")

# performing calculation
if choice == "1":
    print("Result:", add(number1, number2))
elif choice == "2":
    print("Result:", subtract(number1, number2))
elif choice == "3":
    print("Result:", multiply(number1, number2))
elif choice == "4":
    print("Result:", divide(number1, number2))
else:
    print("Invalid choice")
```

LIST AND DICTIONARIES

1. Remove the first duplicate element of a list


```python
numbers = [1, 2, 3, 2, 4, 5]

seen = set()

for i in range(len(numbers)):
    if numbers[i] in seen:
        numbers.pop(i)
        break
    else:
        seen.add(numbers[i])

print(numbers)
```

2. Write a Python program to create a new dictionary by extracting the mentioned keys from the below dictionary.
sample_dict = { "name": "Anderson", "age": 35, "salary": 5000, "city": "London"}
keys = ["name", "salary"]


```python
sample_dict = {
    "name": "Anderson",
    "age": 35,
    "salary": 5000,
    "city": "London"
}

keys = ["name", "salary"]

# create new dictionary with selected keys
new_dict = {}

for key in keys:
    if key in sample_dict:
        new_dict[key] = sample_dict[key]

print(new_dict)
```

3. Write a Python program to convert two lists into a dictionary in a way that item from list_one is the key and item from list_two is the value.
list_one = ['one', 'two', 'three']
list_two = [1, 2, 3]


```python
list_one = ['one', 'two', 'three']
list_two = [1, 2, 3]

# create dictionary by zipping the lists
result_dict = dict(zip(list_one, list_two))

print(result_dict)
```
